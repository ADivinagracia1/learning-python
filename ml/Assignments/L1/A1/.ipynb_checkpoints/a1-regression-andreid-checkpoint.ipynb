```python
#Function to output graphs
def viz_linear(x, y, lin_reg, title, xlbl, ylbl):
    plt.scatter(x.reshape(-1, 1), y, color='red')
    plt.plot(x.reshape(-1, 1), lin_reg.predict(x.reshape(-1, 1)), color='blue')
    plt.title('Linear Regression: ' + title)
    plt.xlabel(xlbl)
    plt.ylabel(ylbl)
    plt.show()
```

```python
#Preprocess train and test figures for NOX
train_nox_x = df_train['AT'].to_numpy()
train_nox_y = df_train['NOX'].to_numpy()
lin_reg_nox.fit(train_nox_x.reshape(-1, 1), train_nox_y)
test_nox_x = df_test['AT'].to_numpy()
test_nox_y = df_test['NOX'].to_numpy()
#lin_reg_nox_test.fit(test_nox_x.reshape(-1, 1), test_nox_y)

#Preprocess train and test figures for CO
train_co_x = df_train['TIT'].to_numpy()
train_co_y = df_train['CO'].to_numpy()
lin_reg_co.fit(train_co_x.reshape(-1, 1), train_co_y)
test_co_x = df_test['TIT'].to_numpy()
test_co_y = df_test['CO'].to_numpy()
#lin_reg_co_test.fit(test_co_x.reshape(-1, 1), test_co_y)

#Preprocess train and test figures for TEY
train_tey_x = df_train['CDP'].to_numpy()
train_tey_y = df_train['TEY'].to_numpy()
lin_reg_tey.fit(train_tey_x.reshape(-1, 1), train_tey_y)
test_tey_x = df_test['CDP'].to_numpy()
test_tey_y = df_test['TEY'].to_numpy()
#lin_reg_tey_test.fit(test_tey_x.reshape(-1, 1), test_tey_y)
```

### Linear Regression - Testing

```python
#Testing NOX with actual and predicted
R_test_nox_x = random.randrange(len(test_nox_x))
rand_test_nox = train_nox_x[R_test_nox_x].reshape(-1, 1) #Actual
y_pred_nox = lin_reg_nox.predict(rand_test_nox)          #Prediction
print('Predicted NOX for sample %d:\t%d' %(R_test_nox_x, y_pred_nox))
print('Actual NOX for sample %d:\t%d' %(R_test_nox_x, train_nox_y[R_test_nox_x]))

#Testing CO actual and predicted
R_test_co_x = random.randrange(len(test_co_x))
rand_test_co = train_co_x[R_test_co_x].reshape(-1, 1) #Actual
y_pred_co = lin_reg_co.predict(rand_test_co)          #Prediction
print('Predicted CO for sample %d:\t%d' %(R_test_co_x, y_pred_co))
print('Actual CO for sample %d:\t%d' %(R_test_co_x, train_co_y[R_test_co_x]))

#Testing CO actual and predicted
R_test_tey_x = random.randrange(len(test_tey_x))
rand_test_tey = train_tey_x[R_test_tey_x].reshape(-1, 1) #Actual
y_pred_tey = lin_reg_tey.predict(rand_test_tey)          #Prediction
print('Predicted TEY for sample %d:\t%d' %(R_test_tey_x, y_pred_tey))
print('Actual TEY for sample %d:\t%d' %(R_test_tey_x, train_tey_y[R_test_tey_x]))
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn import linear_model

def get_model_values(model, x_test, y_test): 
    y_pred = model.predict(x_test.reshape(-1, 1))
    r2     = r2_score(y_test, y_pred)
    rmse   = mean_squared_error(y_test, y_pred, squared = False) 
    return r2, rmse
```

#### R2 and RMSE (Built In)

```python
nox_r2, nox_rmse = get_model_values(lin_reg_nox, test_nox_x, test_nox_y)
co_r2,  co_rmse  = get_model_values(lin_reg_co,  test_co_x,  test_co_y)
tey_r2, tey_rmse = get_model_values(lin_reg_tey, test_tey_x, test_tey_y)
```

```python
print(f"NOX R2:\t{nox_r2}\t\tNOX RMSE:\t{nox_rmse}")
print(f"CO R2:\t{co_r2}\t\tCO RMSE:\t{co_rmse}")
print(f"TEY R2:\t{nox_r2}\t\tTEY RMSE:\t{nox_rmse}")
```

```python
#Self coded RMSE fuction
def my_RMSE(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)**(1/2)

#Function to Plot RMSE values
def rmse_bar_plot(x, y, targetname):
    fig = plt.figure()
    plt.bar(x, y)
    plt.title(f"{targetname} vs RMSE Value")
    plt.xlabel(f"Index in {targetname}")
    plt.ylabel("RMSE Value")
    fig.set_figheight(6)
    fig.set_figwidth(10)
    plt.show()
```

```python
#Output coded RMSE function
print(f"RMSE for NOX:  {my_RMSE(test_nox_y, lin_reg_nox.predict(test_nox_x.reshape(-1, 1)))}")
print(f"RMSE for CO:   {my_RMSE(test_co_y,  lin_reg_co.predict(test_co_x.reshape(-1, 1)))}")
print(f"RMSE for TEY:  {my_RMSE(test_tey_y, lin_reg_tey.predict(test_tey_x.reshape(-1, 1)))}")
```

```python
#Generate RMSE valuse into array
rmse_list_nox = [my_RMSE(y_test, y_pred) for y_pred, y_test in zip(lin_reg_nox.predict(test_nox_x.reshape(-1, 1)), test_nox_y)]
rmse_list_co  = [my_RMSE(y_test, y_pred) for y_pred, y_test in zip(lin_reg_co.predict(test_co_x.reshape(-1, 1)),   test_co_y)]
rmse_list_tey = [my_RMSE(y_test, y_pred) for y_pred, y_test in zip(lin_reg_tey.predict(test_tey_x.reshape(-1, 1)), test_tey_y)]
```

```python
#Generate Bar Graphs (Takes A long time to run)
rmse_bar_plot(range(len(test_nox_x)), rmse_list_nox, 'NOX')    
rmse_bar_plot(range(len(test_co_x)),  rmse_list_co,  'CO')
rmse_bar_plot(range(len(test_tey_x)), rmse_list_tey, 'TEY')
```

Referencing the section "Linear Regression Graphs: Training Set," we see that the correlation coefficient coressponds with the RMSE of each datapoint. 

The dataset for NOX has lowest correlation coefficient and it shows on the RMSE bar graph. Most RMSE values are 20 and below with a few outliers that has a RMSE of around 45. This shows that there is some kind of correlation but it is clear that the AT and the NOX relationship is not very correlated with a coefficient of -0.56. Thus, our model can somewhat predict the trend of this feature.

The dataset for CO is considerably better than the one for NOX. The majority of its RMSE value is under 3, a few go over 5, and only two values with a large error around 20. We can infer that those outliers is what causes the correlation coefficient to be -0.71. Whereas without those datapoints, we could see that our model can better predict the true values of this feature.

The dataset for TEY is the best among the three. There aren't any outliers in the RMSE bar graph. The RMSE value for each datapoint does not go beyond 7 where the majority is much lower. Since the treined set had a correlation coefficient of 0.988, our model had little trouble predicting the actual values for the datapoints. 

# Multiple Linear Regression

```python
from sklearn import linear_model

def get_model_values_M(model, x_test, y_test):
    y_pred = model.predict(x_test)
    r2     = r2_score(y_test, y_pred)
    rmse   = mean_squared_error(y_test, y_pred, squared = False)
    return r2, rmse

mult_reg_tey = linear_model.LinearRegression().fit(df_train[features], df_train['TEY'])
mult_reg_co  = linear_model.LinearRegression().fit(df_train[features], df_train['CO'])
```

```python
#Set up training models
train_co_x_m = df_train[features].to_numpy()
train_co_y_m = df_train['CO'].to_numpy()
test_co_x_m  = df_train[features].to_numpy()
test_co_y_m  = df_train['CO'].to_numpy()
mult_reg_co.fit(train_co_x_m, train_co_y_m)

train_tey_x_m = df_train[features].to_numpy()
train_tey_y_m = df_train['TEY'].to_numpy()
test_tey_x_m  = df_train[features].to_numpy()
test_tey_y_m  = df_train['TEY'].to_numpy()
mult_reg_tey.fit(train_tey_x_m, train_tey_y_m)
```

```python
#Create R2 and RMSE values
co_r2_m,  co_rmse_m  = get_model_values_M(mult_reg_co,  test_co_x_m,  test_co_y_m)
tey_r2_m, tey_rmse_m = get_model_values_M(mult_reg_co,  test_tey_x_m, test_tey_y_m)

#Store RMSE values
rmse_list_co_m  = [my_RMSE(y_test_m, y_pred_m) for y_pred_m, y_test_m in zip(mult_reg_co.predict(test_co_x_m),   test_co_y_m)]
rmse_list_tey_m = [my_RMSE(y_test_m, y_pred_m) for y_pred_m, y_test_m in zip(mult_reg_tey.predict(test_tey_x_m), test_tey_y_m)]
```

```python
#Plot RMSE Values
rmse_bar_plot(range(len(test_co_x_m)),  rmse_list_co_m,  'CO')
rmse_bar_plot(range(len(test_tey_x_m)), rmse_list_tey_m, 'TEY')
```

Examining the graphs between the simple linear regression of CO and the multiple linear regression of CO, the performance is inconclusive. There are slight differences in where the peaks in the RMSE dataset are. It seems that the simple linear regession has a slight advantage but it isn't too significant. Employing more features does not improve the performance of CO.

The comparison for the TEY RMSE dataset is a different story. The scale for the RMSE is halved in the multiple linear regression than the simple linear regression. Thus, multiple features would allow an improvement of performance

# Polynomial Regression

```python
import random
from sklearn import preprocessing
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures

def rmse_bar_plot_poly(x, y, targetname, degree):
    fig = plt.figure()
    plt.bar(x, y)
    plt.title(f"RMSE Value of {targetname} with degree {degree}")
    plt.xlabel(f"Index in {targetname}")
    plt.ylabel("RMSE Value")
    fig.set_figheight(6)
    fig.set_figwidth(10)
    plt.show()
```

```python
poly_train_x = df_train[features]
poly_train_y = df_train['TEY']
poly_test_x = df_test[features]
poly_test_y = df_test['TEY']

degrees = [2,5,11]
```

```python
for degree in degrees:
    scaler = preprocessing.StandardScaler()
    polyreg_scaled = make_pipeline(PolynomialFeatures(degree),scaler,LinearRegression())
    polyreg_scaled.fit(poly_train_x, poly_train_y)
    poly_r2, poly_rmse = get_model_values_M(polyreg_scaled, poly_train_x, poly_train_y)
    rmse_list_poly  = [my_RMSE(y_test, y_pred) for y_pred, y_test in zip(polyreg_scaled.predict(poly_test_x), poly_test_y)]
    rmse_bar_plot_poly(range(len(poly_test_x)), rmse_list_poly, 'TEY', degree)
```
